```python
from torchvision import models
import torch
import torch.nn as nn

model = models.resnext50_32x4d(pretrained=True)
```

```python
from PIL import Image
from torchvision.transforms import ToTensor
```

```python
model.eval()
```

```python
def feature_extractor_fn(image_path):
  with torch.no_grad():
    feature_extractor = torch.nn.Sequential(*list(model.children())[:-1])
    x = Image.open(image_path)
    x = ToTensor()(x).unsqueeze(0)
    output = feature_extractor(torch.tensor(x)) # output now has the features corresponding to input x
    return output.squeeze()[:-1].tolist()
```

```python
features = []
import numpy as np # linear algebra
import pandas as pd 
```

```python
train_path = '/content/drive/MyDrive/522-project/left_0_new/'
```

```python
import os
count = 1
for filename in os.listdir(train_path):
  features1 = ['$'+ str(filename[0:5]), feature_extractor_fn(train_path + filename)]
  # print(str)
  # features1[0] = int(filename[0:5])
  # features2[1] = 
  features.append(features1)



  # features[int(filename[0:5])] = feature_extractor_fn(train_path + filename)

```
