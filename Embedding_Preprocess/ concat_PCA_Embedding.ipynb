```python
df = pd.concat( map(pd.read_csv, image_embedding_paths), ignore_index=True)
df.head()
```

```python
df.drop_duplicates(subset="0",keep='first',inplace=True)

# a = df.groupby("0").count().sort_values(by="1", ascending=False)
# print(a)
# print("&&&&&&&&&&&")
```

```python
df["0"] = df["0"].str.replace("$", "").astype(int)
```

```python
text_train_embeddings = pd.read_csv("/content/drive/MyDrive/522-project/Embeddings/Roberta_retrain150/retrain_embedding_1_train.csv")
text_test_unseen_embeddings = pd.read_csv("/content/drive/MyDrive/522-project/Embeddings/Roberta_retrain150/retrain_embedding_1_test_unseen.csv")
text_test_seen_embeddings = pd.read_csv("/content/drive/MyDrive/522-project/Embeddings/Roberta_retrain150/retrain_embedding_1_test_seen.csv")
text_dev_unseen_embeddings = pd.read_csv("/content/drive/MyDrive/522-project/Embeddings/Roberta_retrain150/retrain_embedding_1_val.csv")
```

```python
text_train_embeddings.shape, text_train_embeddings.head()
```

```python
merged_train = pd.merge(text_train_embeddings, df, how='inner', left_on="0", right_on="0", suffixes=('_text', '_image'))

merged_train_check = pd.merge(text_train_embeddings, df, how='outer', left_on="0", right_on="0", suffixes=('_text', '_image'), indicator=True)
merged_train_check[merged_train_check._merge =="left_only"]
```

```python
#test_unseen
merged_test_unseen = pd.merge(text_test_unseen_embeddings, df, how='inner', left_on="0", right_on="0", suffixes=('_text', '_image'))

merged_test_unseen_check = pd.merge(text_test_unseen_embeddings, df, how='outer', left_on="0", right_on="0", suffixes=('_text', '_image'), indicator=True)
merged_test_unseen_check[merged_test_unseen_check._merge =="left_only"]
```

```python
#test_seen
merged_test_seen = pd.merge(text_test_seen_embeddings, df, how='inner', left_on="0", right_on="0", suffixes=('_text', '_image'))

merged_test_seen_check = pd.merge(text_test_seen_embeddings, df, how='outer', left_on="0", right_on="0", suffixes=('_text', '_image'), indicator=True)
merged_test_seen_check[merged_test_seen_check._merge =="left_only"]
```

```python
#dev_unseen
merged_dev_unseen = pd.merge(text_dev_unseen_embeddings, df, how='inner', left_on="0", right_on="0", suffixes=('_text', '_image'))

merged_dev_unseen_check = pd.merge(text_dev_unseen_embeddings, df, how='outer', left_on="0", right_on="0", suffixes=('_text', '_image'), indicator=True)
merged_dev_unseen_check[merged_dev_unseen_check._merge =="left_only"]
```

```python
import json
##train
merged_train["combine_text"] = merged_train["combine_text"].apply(json.loads)
merged_train["combine_image"] = merged_train["combine_image"].apply(json.loads)

##train
merged_dev_unseen["combine_text"] = merged_dev_unseen["combine_text"].apply(json.loads)
merged_dev_unseen["combine_image"] = merged_dev_unseen["combine_image"].apply(json.loads)

##train
merged_test_seen["combine_text"] = merged_test_seen["combine_text"].apply(json.loads)
merged_test_seen["combine_image"] = merged_test_seen["combine_image"].apply(json.loads)

##train
merged_test_unseen["combine_text"] = merged_test_unseen["combine_text"].apply(json.loads)
merged_test_unseen["combine_image"] = merged_test_unseen["combine_image"].apply(json.loads)

```

```python
merged_train["concat_embeddings"] = merged_train["combine_text"] +  merged_train["combine_image"]

merged_dev_unseen["concat_embeddings"] = merged_dev_unseen["combine_text"] +  merged_dev_unseen["combine_image"]

merged_test_seen["concat_embeddings"] = merged_test_seen["combine_text"] +  merged_test_seen["combine_image"]

merged_test_unseen["concat_embeddings"] = merged_test_unseen["combine_text"] +  merged_test_unseen["combine_image"]
```

```python
merged_train[["concat_embeddings", "0"]].to_csv("/content/drive/MyDrive/522-project/Embeddings/Retrain-Concat-150T-50I/ConcatEmbeddings_1_train.csv")

merged_dev_unseen[["concat_embeddings", "0"]].to_csv("/content/drive/MyDrive/522-project/Embeddings/Retrain-Concat-150T-50I/ConcatEmbeddings_1_dev_unseen.csv")
merged_test_seen[["concat_embeddings", "0"]].to_csv("/content/drive/MyDrive/522-project/Embeddings/Retrain-Concat-150T-50I/ConcatEmbeddings_1_test_seen.csv")
merged_test_unseen[["concat_embeddings", "0"]].to_csv("/content/drive/MyDrive/522-project/Embeddings/Retrain-Concat-150T-50I/ConcatEmbeddings_1_test_unseen.csv")




```
